## Introduction

Weather forcasting is one of the most commonly used applications of machine learning. Often times current or past weather patterns can give us a clue on what the weather is going to be like in the future. While modern day weather networks use state-of-the-art data collection systems and Neural Networks to predict up to a week's worth of wather patterns, we (with our limited data) are going to predict whether or not it is going to rain tomorrow.

## Data Preprocessing

The first step in any data science-related task is to know our data. Here we will load up the `weatherAUS.csv` dataset that contains the weather data from towns and cities in Australia. Upon inspecting the data, we will choose our features based on what we think we need, as well as what the weather API is able to provide for us.

```python
import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)

dataset = pd.read_csv("datasets/weatherAUS.csv")
dataset = dataset.sample(frac=1).reset_index(drop=True)
dataset
```

The first step is to convert the columns `"RainToday"` and `"RainTomorrow"` into binary labels.

```python
rain_dummies = pd.get_dummies(dataset[['RainToday', 'RainTomorrow']])
dataset['rain_today_num'] = rain_dummies[["RainToday_Yes"]]
dataset['rain_tomorrow_num'] = rain_dummies[["RainTomorrow_Yes"]]
dataset.head(5)
```

```python
items = ['Pressure3pm', 'WindSpeed3pm', 'MinTemp', 'MaxTemp', 'Humidity3pm', 'Cloud3pm', 'rain_today_num']

for i in items:
    col_np = dataset[i].to_numpy()
    col_np = np.nan_to_num(col_np, nan=0)
    
    non_zero = np.count_nonzero(~np.isnan(col_np))
    mean = np.sum(col_np)/non_zero
    
    dataset[[i]] = dataset[[i]].fillna(mean)
    print(f"{i} mean: {mean}")
    
dataset
```

```python
features = dataset[items].to_numpy()
labels = dataset[['rain_tomorrow_num']].to_numpy()

print(f"features shape: {features.shape}")
print(f"labels shape: {labels.shape}", '\n')
```

## Training the Classifier



```python
import torch
from torch.nn import Module, Linear, BCELoss, BatchNorm1d
from torch.optim import SGD
from torch.autograd import Variable

class Logreg(Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.linear  = Linear(input_dim, output_dim)
        self.bn = BatchNorm1d(input_dim)
        
    def forward(self, x):
        x = self.bn(x)
        z = self.linear(x)
        y_pred = torch.sigmoid(z)
        return y_pred
    
    def fit(self, X_train, Y_train, epochs, loss_func, opt):
        for i in range(epochs):
            self.train()
            opt.zero_grad()
        
            Y_pred = self(X_train)
            loss = loss_func(Y_pred, Y_train)
            
            print(f"iteration {i+1}: loss: {loss.item()}")
        
            loss.backward()
            opt.step()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

X_train, X_test, Y_train, Y_test = train_test_split(features, labels, test_size=0.1)

X_train = poly.fit_transform(X_train)
X_test = poly.fit_transform(X_test)

X_train_t = torch.from_numpy(X_train).float()
Y_train_t = torch.from_numpy(Y_train).float()

print(f"X_train shape: {X_train.shape}")
print(f"Y_train shape: {Y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Y_test shape: {Y_test.shape}")
```

```python
LEARNING_RATE = 0.3
EPOCHS = 300

model = Logreg(X_train.shape[1], 1)

criterion = BCELoss()
optimizer = SGD(model.parameters(), lr=LEARNING_RATE, momentum=0.9, nesterov=True)

if __name__ == "__main__":
    model.fit(X_train_t, Y_train_t, EPOCHS, criterion, optimizer)
```

```python
from sklearn.metrics import confusion_matrix, f1_score
import seaborn as sn

X_test_t = torch.from_numpy(X_test).float()
Y_test_t = torch.from_numpy(Y_test).float()

model.eval()

Y_pred_t = model(X_test_t)
Y_pred = (Y_pred_t.detach().numpy() > 0.5).astype(np.int32)

scores = f1_score(Y_test, Y_pred)
accuracy = np.sum(Y_pred == Y_test)/Y_test.size

print(f"F1 Score: {scores}")
print(f"Accuracy: {accuracy}")
```

```python
def plot_conf_mat(Y_pred, Y_gt):
    conf_mat = pd.DataFrame(confusion_matrix(Y_gt, Y_pred, labels=[0,1]))
    sn.heatmap(conf_mat, annot=True)
    
plot_conf_mat(Y_pred, Y_test)
```

```python
if __name__ == "__main__":
    torch.save(model.state_dict(), "rain_predictor.pt")
```
