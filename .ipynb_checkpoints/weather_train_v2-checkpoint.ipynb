```python
import numpy as np
import pandas as pd
import glob

pd.set_option('display.max_columns', None)

base_path = 'datasets/weather/'
dataset_raw = pd.DataFrame()

for i in glob.iglob(base_path + '*.csv'):
    dataframe_section = pd.read_csv(i)
    print(f"{i} shape: {dataframe_section.shape}")
        
```

```python
feature_names = ['temp', 'pressure', 'weather_description', 'wind_speed', 'wind_dir', 'humidity']

for i, path in enumerate(glob.iglob(base_path + '*.csv')):
    dataframe_section = pd.read_csv(path)
    column = np.array([])
    if dataframe_section.shape[1] == 37:
        for j in range(1, len(dataframe_section.columns)):
            column = np.concatenate((column, dataframe_section.iloc[:, j]))
        dataset_raw[feature_names[i]] = column.flatten()
        column = np.array([])

dataset_raw
```

```python
dataset = dataset_raw.iloc[:600000, :]
weather_desc = pd.get_dummies(dataset['weather_description'])
weather_desc
```

```python
dataset['rain'] = weather_desc['drizzle'] | weather_desc['freezing rain'] | weather_desc['heavy intensity drizzle'] | weather_desc['heavy intensity rain'] | weather_desc['heavy intensity shower rain'] |  weather_desc['light intensity drizzle'] | weather_desc['light intensity drizzle rain'] | weather_desc['light intensity shower rain'] | weather_desc['light rain'] | weather_desc['light rain and snow'] |  weather_desc['proximity moderate rain'] | weather_desc['proximity shower rain'] | weather_desc['proximity thunderstorm'] | weather_desc['proximity thunderstorm with drizzle'] | weather_desc['proximity thunderstorm with rain'] | weather_desc['ragged thunderstorm'] | weather_desc['shower rain'] | weather_desc['thunderstorm'] | weather_desc['thunderstorm with drizzle'] | weather_desc['thunderstorm with heavy drizzle'] | weather_desc['thunderstorm with heavy rain'] | weather_desc['thunderstorm with light drizzle'] | weather_desc['thunderstorm with light rain'] | weather_desc['thunderstorm with rain'] | weather_desc['very heavy rain']
```

```python
feature_names = ['temp', 'pressure', 'wind_speed', 'humidity', 'rain']

for i in feature_names:
    column = dataset[i].to_numpy()
    num_count = np.count_nonzero(~np.isnan(column))
    column = np.nan_to_num(column, nan=0)
    mean = np.sum(column)/num_count
    dataset.loc[:, i] = dataset.loc[:,i].fillna(mean)
    
dataset
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

features = dataset[['temp', 'pressure', 'wind_speed', 'humidity']].to_numpy()
labels = dataset[['rain']].to_numpy()

X_train, X_test, Y_train, Y_test = train_test_split(features, labels, test_size = 0.8)
poly = PolynomialFeatures(degree=2)

X_train = poly.fit_transform(X_train)
X_test = poly.fit_transform(X_test)


print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
positives = np.sum(dataset['rain'].to_numpy())
negatives = int(len(dataset.index) - positives)

print(f'Num of positives: {positives}')
print(f'Num of negatives: {negatives}\n')

pos_weight = len(dataset.index)/(2 * positives)
neg_weight = len(dataset.index)/(2 * negatives)

print(f"positive weight: {pos_weight}")
print(f"negative weight: {neg_weight}")

def class_weights(Y_train, pos_weight, neg_weight):
    pos_mask = Y_train.astype(bool)
    neg_mask = ~Y_train.astype(bool)

    pos_mask = pos_mask.astype(np.float32) * pos_weight

    neg_mask = neg_mask.astype(np.float32) * neg_weight
    return pos_mask + neg_mask
    

weights = class_weights(Y_train, pos_weight, neg_weight)
print(weights.shape)
```

```python
import torch
from torch.nn import Module, Linear, BCELoss, BatchNorm1d
from torch.optim import SGD, Adam
from torch.autograd import Variable

class Logreg(Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.linear  = Linear(input_dim, output_dim)
        self.bn = BatchNorm1d(input_dim)
        
        self.inputbn = BatchNorm1d(input_dim)
        self.linear1 = Linear(input_dim, 100)
        self.bn1 = BatchNorm1d(100)
        self.linear2 = Linear(100, output_dim)
        
#     def forward(self, x):
#         x = self.bn(x)
#         z = self.linear(x)
#         y_pred = torch.sigmoid(z)
#         return y_pred

    def forward(self, x):
        x = self.inputbn(x)
        x = self.linear1(x)
        x = self.bn1(x)
        x = torch.relu(x)
        
        x = self.linear2(x)
        x = torch.sigmoid(x)
        return x
        
    
    def fit(self, X_train, Y_train, epochs, loss_func, opt):
        for i in range(epochs):
            self.train()
            opt.zero_grad()
        
            Y_pred = self(X_train)
            loss = loss_func(Y_pred, Y_train)
            
            print(f"iteration {i+1}: loss: {loss.item()}")
        
            loss.backward()
            opt.step()
```

```python
LEARNING_RATE = 0.5
EPOCHS = 1000

model = Logreg(X_train.shape[1], 1)
model = model.cuda()

criterion = BCELoss(weight=torch.from_numpy(weights))
optimizer = Adam(model.parameters())

X_train_t = torch.from_numpy(X_train).float().cuda()
Y_train_t = torch.from_numpy(Y_train).float().cuda()


if __name__ == "__main__":
    model.fit(X_train_t, Y_train_t, EPOCHS, criterion, optimizer)
```

```python
from sklearn.metrics import confusion_matrix, f1_score
import seaborn as sn

X_test_t = torch.from_numpy(X_test).float()
Y_test_t = torch.from_numpy(Y_test).float()

model.eval()

Y_pred_t = model(X_test_t)
Y_pred = (Y_pred_t.detach().numpy() > 0.5).astype(np.int32)

scores = f1_score(Y_test, Y_pred)
accuracy = np.sum(Y_pred == Y_test)/Y_test.size

print(f"F1 Score: {scores}")
print(f"Accuracy: {accuracy}")
```

```python
def plot_conf_mat(Y_pred, Y_gt):
    conf_mat = pd.DataFrame(confusion_matrix(Y_gt, Y_pred, labels=[0,1]))
    sn.heatmap(conf_mat, annot=True)
    
plot_conf_mat(Y_pred, Y_test)
```
